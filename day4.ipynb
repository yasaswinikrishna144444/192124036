```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
std_dev_men = [4, 3, 4, 1, 5]
std_dev_women = [3, 5, 2, 3, 3]

# Create an array of X values for the groups
x = np.arange(len(groups))

# Create a stacked bar plot
plt.bar(x, means_men, yerr=std_dev_men, label='Men', color='blue', capsize=5)
plt.bar(x, means_women, yerr=std_dev_women, bottom=means_men, label='Women', color='orange', capsize=5)

# Add labels for the x-axis and y-axis
plt.xlabel('Groups')
plt.ylabel('Scores')

# Set the X-axis labels to the group names
plt.xticks(x, groups)

# Add a title to the plot
plt.title('Stacked Bar Plot with Error Bars')

# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for X and Y
np.random.seed(0)  # For reproducibility
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)

# Create a scatter plot with empty circles
plt.scatter(x, y, color='blue', marker='o', facecolors='none')

# Add labels for the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title to the plot
plt.title('Scatter Plot with Empty Circles')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for ball positions (X, Y) and sizes
np.random.seed(0)  # For reproducibility
num_balls = 50
x = np.random.rand(num_balls)
y = np.random.rand(num_balls)
sizes = np.random.rand(num_balls) * 100

# Create a scatter plot with balls of different sizes
plt.scatter(x, y, s=sizes, c='blue', alpha=0.7, edgecolors='black', linewidths=1)

# Add labels for the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title to the plot
plt.title('Scatter Plot with Balls of Different Sizes')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for three groups (Weights and Heights)
group1_weights = np.random.normal(150, 10, 50)
group1_heights = np.random.normal(65, 5, 50)

group2_weights = np.random.normal(160, 12, 50)
group2_heights = np.random.normal(68, 6, 50)

group3_weights = np.random.normal(170, 15, 50)
group3_heights = np.random.normal(70, 7, 50)

# Create a scatter plot for each group
plt.scatter(group1_weights, group1_heights, color='red', label='Group 1', marker='o')
plt.scatter(group2_weights, group2_heights, color='blue', label='Group 2', marker='s')
plt.scatter(group3_weights, group3_heights, color='green', label='Group 3', marker='^')

# Add labels for the x-axis and y-axis
plt.xlabel('Weights')
plt.ylabel('Heights')

# Add a title to the plot
plt.title('Scatter Plot of Weights and Heights for Three Groups')

# Add a legend
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
