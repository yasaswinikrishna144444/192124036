```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample stock price data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'Stock Price': [100, 105, 110, 115]
}

# Create a DataFrame
stock_data = pd.DataFrame(data)

# Set the 'Date' column as the index if not already
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace=True)

# Select data between two specific dates
start_date = '2023-01-01'
end_date = '2023-01-04'

selected_data = stock_data[(stock_data.index >= start_date) & (stock_data.index <= end_date)]

# Create a line plot
plt.figure(figsize=(12, 6))
plt.plot(selected_data.index, selected_data['Stock Price'], label='Alphabet Inc. Stock Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Alphabet Inc. Stock Price (January 2023)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample trading volume data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'Trading Volume': [10000, 11000, 9500, 10500]
}

# Create a DataFrame
stock_data = pd.DataFrame(data)

# Set the 'Date' column as the index if not already
stock_data['Date'] = pd.to_datetime(stock_data['Date'])
stock_data.set_index('Date', inplace=True)

# Select data between two specific dates
start_date = '2023-01-01'
end_date = '2023-01-04'

selected_data = stock_data[(stock_data.index >= start_date) & (stock_data.index <= end_date)]

# Create a bar plot for trading volume
plt.figure(figsize=(12, 6))
plt.bar(selected_data.index, selected_data['Trading Volume'], color='green', label='Trading Volume')
plt.xlabel('Date')
plt.ylabel('Trading Volume')
plt.title('Alphabet Inc. Trading Volume (January 2023)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for stock prices and trading volume
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'Stock Price': [100, 105, 110, 115],
    'Trading Volume': [10000, 11000, 9500, 10500]
}

# Create a DataFrame
stock_data = pd.DataFrame(data)

# Create a scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(stock_data['Stock Price'], stock_data['Trading Volume'], c='blue', alpha=0.7)
plt.xlabel('Stock Price')
plt.ylabel('Trading Volume')
plt.title('Alphabet Inc. Stock Price vs Trading Volume')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Sample data for sales
data = {
    'Item': ['Television', 'Home Theater', 'Television', 'Cell Phone'],
    'Sale_amt': [113810.00, 25000.00, 43128.00, 6075.00]
}

# Create a DataFrame
sales_data = pd.DataFrame(data)

# Create a Pivot table to find the maximum and minimum sale value of items
pivot_table = pd.pivot_table(sales_data, values='Sale_amt', index='Item', aggfunc=[max, min])

print(pivot_table)
```

```python
import pandas as pd

# Sample data for sales
data = {
    'Item': ['Television', 'Home Theater', 'Television', 'Cell Phone'],
    'Units': [95, 50, 36, 27]
}

# Create a DataFrame
sales_data = pd.DataFrame(data)

# Create a Pivot table to find item-wise unit sold
pivot_table = pd.pivot_table(sales_data, values='Units', index='Item', aggfunc='sum')

print(pivot_table)
```

```python
import pandas as pd

# Sample data for sales
data = {
    'Region': ['East', 'Central', 'Central', 'Central'],
    'Manager': ['Martha', 'Hermann', 'Hermann', 'Timothy'],
    'SalesMan': ['Alexander', 'Shelli', 'Luis', 'David'],
    'Sale_amt': [113810.00, 25000.00, 43128.00, 6075.00]
}

# Create a DataFrame
sales_data = pd.DataFrame(data)

# Create a Pivot table to find the total sale amount by Region, Manager, and SalesMan
pivot_table = pd.pivot_table(sales_data, values='Sale_amt', index=['Region', 'Manager', 'SalesMan'], aggfunc='sum')

print(pivot_table)
```

```python
import pandas as pd
import numpy as np
from IPython.display import HTML

# Create a DataFrame with random values
data = pd.DataFrame(np.random.randint(-10, 10, size=(10, 4)), columns=['A', 'B', 'C', 'D'])

# Function to highlight negative numbers in red and positive numbers in black
def highlight_numbers(val):
    color = 'red' if val < 0 else 'black'
    return f'color: {color}'

# Apply the style to the DataFrame
styled_data = data.style.applymap(highlight_numbers)

# Display the styled DataFrame
HTML(styled_data.render())
```
